```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def OA(v, v_hat, pitch_correct):
    return (v*v_hat*pitch_correct) + (1-v)*(1-v_hat)

def RPA(v, v_hat, pitch_correct):
    return (v * pitch_correct)

def VA(v, v_hat):
    return (v * v_hat) + (1-v)*(1-v_hat)
```

```python
v_grid = np.linspace(0, 1, 1000)
plt.plot(v_grid, '.')
```

```python
def compute_metric(v_vals, v_hat_vals):
    OA_grid = -1.0 * np.ones((1000, 1000))
    OA_grid_incorrect = -1.0 * np.ones((1000, 1000))
    RPA_grid = -1.0 * np.ones((1000, 1000))
    RPA_grid_incorrect = -1.0 * np.ones((1000, 1000))
    VA_grid = -1.0 * np.ones((1000, 1000))
    for i, v in enumerate(v_vals):
        for j, v_hat in enumerate(v_hat_vals):
            OA_grid[i, j] = OA(v, v_hat, 1.0)
            OA_grid_incorrect[i, j] = OA(v, v_hat, 0.0)
            RPA_grid[i, j] = RPA(v, v_hat, 1.0)
            RPA_grid_incorrect[i, j] = RPA(v, v_hat, 0.0)
            VA_grid[i, j] = VA(v, v_hat)

    return OA_grid, OA_grid_incorrect, RPA_grid, RPA_grid_incorrect, VA_grid
```

```python
v_vals_continuous = np.linspace(0, 1, 1000)
v_vals_binary = np.round(v_vals_continuous)

OA_cont, OA_cont_i, RPA_cont, RPA_cont_i, VA_cont = compute_metric(
    v_vals_continuous, v_vals_continuous)

OA_bin_cont, OA_bin_cont_i, RPA_bin_cont, RPA_bin_cont_i, VA_bin_cont = compute_metric(
    v_vals_binary, v_vals_continuous)

OA_cont_bin, OA_cont_bin_i, RPA_cont_bin, RPA_cont_bin_i, VA_cont_bin = compute_metric(
    v_vals_continuous, v_vals_binary)

OA_bin, OA_bin_i, RPA_bin, RPA_bin_i, VA_bin = compute_metric(
    v_vals_binary, v_vals_binary)
```

```python
plt.figure(figsize=(20, 5))

plt.suptitle("Overall Accuracy: Correct Pitch")

plt.subplot(131)
plt.title("Continuous Estimate and Label")
plt.imshow(OA_cont, origin='lower', cmap='magma', interpolation='none')
plt.colorbar()
plt.axis('tight')
plt.xticks(np.arange(0, 1200, 200), np.round(np.arange(0, 1.2, 0.2), 1))
plt.yticks(np.arange(0, 1200, 200), np.round(np.arange(0, 1.2, 0.2), 1))
plt.ylabel("Reference voicing")
plt.xlabel("Estimate voicing")

plt.subplot(132)
plt.title("Continuous Estimate, Binary Label")
plt.imshow(OA_bin_cont, origin='lower', cmap='magma', interpolation='none')
plt.colorbar()
plt.axis('tight')
plt.xticks(np.arange(0, 1200, 200), np.round(np.arange(0, 1.2, 0.2), 1))
plt.yticks(np.arange(0, 1200, 200), np.round(np.arange(0, 1.2, 0.2), 1))
plt.ylabel("Reference voicing")
plt.xlabel("Estimate voicing")

plt.subplot(133)
plt.title("Binary Estimate and Label")
plt.imshow(OA_bin, origin='lower', cmap='magma', interpolation='none')
plt.colorbar()
plt.axis('tight')
plt.xticks(np.arange(0, 1200, 200), np.round(np.arange(0, 1.2, 0.2), 1))
plt.yticks(np.arange(0, 1200, 200), np.round(np.arange(0, 1.2, 0.2), 1))
plt.ylabel("Reference voicing")
plt.xlabel("Estimate voicing")

plt.show()
```
